```python
# Importing Necessary Libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
#Loading the Data
data= pd.read_csv('tested.csv')
```

```python
#Displays First 5 Rows in the Dataset
data.head()
```

```python
# Displays last 5 rows in the Dataset
data.tail()
```

```python
# Gives information about the dataset
data.info()
```

```python
# Checking the missing values in the dataset
data.isna().sum()
```

```python
# We have found that most of the values are missing in the Cabin Column so we are dropping that column
data=data.drop(['Cabin'],axis = 1)
```

```python
# Replacing missing values in Age and Fare columns with mean value
data['Age'].fillna(data['Age'].mean(), inplace = True)
data['Fare'].fillna(data['Fare'].mean(),inplace = True)
```

```python
# Displays the total number of rows and columns
data.shape
```

```python
data.isna().sum() # No null values
```

```python
# Getting Statistical information
data.describe()
```

```python
# converting categorical columns into numerical columns
data.replace({'Sex':{'male':0,'female':1}, 'Embarked':{'S':0,'C':1,'Q':2}}, inplace=True)
data.head()
```

```python
# Dividing X and Y parameters
X= data.drop(['PassengerId','Name','Ticket','Survived'], axis =1)
y = data['Survived']
```

```python
print(X)
```

```python
print(y)
```

```python
# Splitting Training and Testing Data
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test= train_test_split(X,y,test_size=0.3,random_state = 42)
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
# Model selection
from sklearn.linear_model import LogisticRegression
lr_model = LogisticRegression()
```

```python
lr_model.fit(X_train, y_train)
```

```python
# Predictions
lr_predic = lr_model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
```

```python
print("accuracy is:",accuracy_score(y_test, lr_predic))
```

```python
print(classification_report(y_test, lr_predic))
```

```python
con=confusion_matrix(y_test, lr_predic)
con
```

We have used Logistic Regression Model which gives 100 % accuracy.It shows that Logistic Regression Model is very accurate.
